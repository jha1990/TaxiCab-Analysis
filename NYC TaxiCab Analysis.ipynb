# Average tip received by a vendors during the day

# Which time of the day is the busiest?

```python
st=list(df1.tpep_pickup_datetime)# Extracting pickup date time which is string format.
ed=list(df1.tpep_dropoff_datetime)# Extracting dropoff date time which is string format.
st_lis=[]# creating empty list
ed_lis=[]# creating empty list
for i in range(len(st)): # iterating over the pick up column and assigning each datetime  to variable tim.
    tim=str(st[i])# in tim variable it has string length of 20.  
    tim1=str(ed[i])
    st_lis.append(tim[11:16])#ex 0 to 10 has location for date and 11 to 16 is the location for time.00:00 format
    ed_lis.append(tim1[11:16])
uni_tim=list(set(st_lis))
max_count=[]
for i in uni_tim:# we are counting busy time, by counting which time period is repeated for maimum
    c=st_lis.count(i)
    max_count.append(c)
v=max(max_count)
busy_time=st_lis[max_count.index(v)]

uni_tim1=list(set(ed_lis))
max_count1=[]
for i in uni_tim1:
    c=ed_lis.count(i)
    max_count1.append(c)
v1=max(max_count1)
busy_time1=ed_lis[max_count1.index(v1)]
print("The busiest time of the day is : ",busy_time," & ",busy_time1)
```
